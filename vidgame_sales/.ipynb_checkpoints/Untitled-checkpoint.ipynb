```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
path = r"data\vgsales.csv"
```

```python
df = pd.read_csv(path)
```

```python
df_clean = df.drop(columns=['Year','Publisher','Platform'])
```

```python
df_clean.head()
```

```python
categorical_cols = ['Genre']
```

```python
df_clean[categorical_cols] = df_clean[categorical_cols].astype('category')
df_clean = pd.get_dummies(df_clean, columns=categorical_cols, drop_first=True)
```

```python
print(df_clean.dtypes)
```

```python
df_clean.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Genre'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Genre')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
cols_to_convert = [
    'NA_Sales',
    'EU_Sales',
    'JP_Sales','Other_Sales',
    'Global_Sales',
    'Genre_Adventure',
    'Genre_Fighting',
    'Genre_Misc',
    'Genre_Platform',
    'Genre_Puzzle',
    'Genre_Racing',
    'Genre_Role-Playing',
    'Genre_Shooter',
    'Genre_Simulation',
    'Genre_Sports',
    'Genre_Strategy'
]

correlation_matrix = df_clean[cols_to_convert].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

Y = df_clean[['Global_Sales']]
X = df_clean.drop(columns=['Name','Global_Sales'])
model = LinearRegression()
model.fit(X, Y)

# Predict
y_pred = model.predict(X)

print("Intercept:", model.intercept_[0])
print("Coefficients:")
print(pd.Series(model.coef_[0], index=X.columns))

# Optional: model score
print("\nR2 score:", model.score(X, Y))
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
model.fit(X_scaled, Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y)


from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

yhat = model.predict(X_test)
```
